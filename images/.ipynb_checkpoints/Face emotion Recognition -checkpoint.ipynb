## Importing Modules

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2
import os
import seaborn as sns
import math
import itertools
# TensorFlow imports
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
from tensorflow.keras.models import Sequential
from tensorflow.keras import models, regularizers
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization, Activation
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint
# scikit-learn imports
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils import class_weight
```

## Loading the images and preprocessing them


```python
train_dir = "train"
valid_dir = "test"

train_datagen = ImageDataGenerator(
    rescale = 1. / 255,
    width_shift_range = 0.1,
    height_shift_range = 0.1,
    rotation_range = 20,
    zoom_range = 0.2,
    horizontal_flip = True,
    shear_range = 0.1,
    fill_mode = 'nearest'
)

validation_datagen = ImageDataGenerator(rescale = 1. / 255)

```

Loading from the directory and making the size of image to 56 pixcels along with augmentation

```python
train_gen = train_datagen.flow_from_directory(
         train_dir,
         target_size = (56,56),
         batch_size = 128,
         class_mode = 'categorical',
         color_mode = 'grayscale',
         shuffle = True
)

valid_gen = validation_datagen.flow_from_directory(
        valid_dir,
        target_size = (56,56),
        batch_size = 128,
        class_mode = 'categorical',
        color_mode = 'grayscale',
        shuffle=False
)
```

## Visulaizing Some images 

```python
class_indices = train_gen.class_indices
# this returns the class name to index
idx_class = {v: k for k, v in class_indices.items()} 
# reversing them to access through the index
seen = set() 
img_per_class = {}

while len(seen) < len(class_indices):
    images, labels = next(train_gen)
    class_index = np.argmax(labels[0])
    if class_index not in seen:
        seen.add(class_index)
        img_per_class[class_index] = images[0]

num_classes = len(class_indices)
cols = 2
rows = (num_classes // cols) + 1

plt.figure(figsize=(cols * 2, rows * 3))
for i, (class_index, img) in enumerate(sorted(img_per_class.items())):
    plt.subplot(rows, cols, i + 1)
    plt.imshow(img.squeeze(), cmap='gray')
    plt.title(idx_class[class_index])
    plt.axis('off')

plt.tight_layout()
plt.show()
```

## CNN Architecture

```python
nb_classes = 7

# Initialising the CNN
model = Sequential()

# 1st Convolution layer
model.add(Conv2D(64,(3,3), padding='same', input_shape=(56, 56,1)))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

# 2nd Convolution layer
model.add(Conv2D(128,(5,5), padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

# 3rd Convolution layer
model.add(Conv2D(512,(3,3), padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

# 4th Convolution layer
model.add(Conv2D(512,(3,3), padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

# Flattening
model.add(Flatten())

# Fully connected layer 1st layer
model.add(Dense(256))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Dropout(0.25))

# Fully connected layer 2nd layer
model.add(Dense(512))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(Dropout(0.25))

model.add(Dense(nb_classes, activation='softmax'))

# Optimizing
opt = Adam(learning_rate=0.0001)
model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])

# Summary
model.summary()
```

## Model Training

```python
epochs = 50
checkpoint = ModelCheckpoint("model_weights.h5", monitor='val_accuracy', verbose=1, save_best_only=True, mode='max')
callbacks_list = [checkpoint]

history = model.fit(
                    train_gen,
                    steps_per_epoch=train_gen.n//train_gen.batch_size,
                    epochs=epochs,
                    validation_data = valid_gen,
                    validation_steps = valid_gen.n//valid_gen.batch_size,
                    callbacks=callbacks_list
                    )
```

## Plotting the accuracy and losses 

```python
train_acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
train_loss = history.history['loss']
val_loss = history.history['val_loss']

# to get best epoch result
best_epoch = val_acc.index(max(val_acc)) + 1
best_acc =  max(val_acc)
print('Best Validation Accuracy Score {:0.5f}, is for epoch {}'.format( best_acc, best_epoch))

# Plotting the results
epochs = range(1, len(train_acc) + 1)
plt.figure(figsize = (6,6))
plt.subplot(2,1,1)
plt.plot(epochs,train_acc,'g', label = 'Training Accuracy')
plt.plot(epochs,val_acc,'r', label = 'Validation Accuracy')
plt.grid(True)
plt.legend()
plt.xlabel('Epoch')

plt.subplot(2,1,2)
plt.plot(epochs, train_loss, 'g', label = 'Training Loss')
plt.plot(epochs, val_loss, 'r', label = 'Validation Loss')
plt.grid(True)
plt.legend()
plt.xlabel('Epoch')
plt.show()
```

## Connecting with Open-Cv for Webcam application

```python
# 1) Load your trained FER model
model = load_model('model_weights.h5')

# 2) Emotion labels
emotions = ['Angry', 'Disgust','Fear','Happy','Sad','Surprise','Neutral']

# 3) Grab one frame from webcam
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise RuntimeError("Could not open webcam")

ret, frame = cap.read()
cap.release()  # release immediately
if not ret:
    raise RuntimeError("Failed to capture image")

# 4) Detect faces
face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
)
gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(
    gray_frame,
    scaleFactor=1.1,
    minNeighbors=5,
    minSize=(48, 48)
)

if len(faces) == 0:
    print("No faces detected.")
else:
    # Process each face
    for (x, y, w, h) in faces:
        face_img = gray_frame[y:y+h, x:x+w]
        face_resized = cv2.resize(face_img, (48, 48)).astype('float32') / 255.0
        face_input = np.expand_dims(face_resized, axis=(0, -1))  # shape (1,48,48,1)

        # 5) Predict emotion
        preds = model.predict(face_input)
        idx = np.argmax(preds[0])
        label = emotions[idx]
        confidence = preds[0][idx]

        # 6) Print to console
        print(f"Predicted emotion: {label} (confidence: {confidence:.2f})")

        # 7) Annotate frame
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0,255,0), 2)
        cv2.putText(frame, f"{label} {confidence:.2f}",
                    (x, y-10), cv2.FONT_HERSHEY_SIMPLEX,
                    0.8, (0,255,0), 2)

# 8) Save & display result
output_path = 'fer_snapshot_result.jpg'
cv2.imwrite(output_path, frame)
print(f"Saved annotated image to {output_path}")

cv2.imshow('FER Snapshot', frame)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Uploading Image and predicting on it

```python
from tkinter import Tk
from tkinter.filedialog import askopenfilenames

Tk().withdraw()

file_paths = askopenfilenames(
    title='Select image files for FER',
    filetypes=[('Image Files', '*.jpg *.jpeg *.png *.bmp')]
)
if not file_paths:
    print("No files selected. Exiting.")
    exit()

# 3) Face detector setup
face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
)

# 4) Process each selected image
for img_path in file_paths:
    # a) Read image
    frame = cv2.imread(img_path)
    if frame is None:
        print(f"Failed to load image: {img_path}")
        continue

    # b) Detect faces in grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(
        gray, scaleFactor=1.1, minNeighbors=5, minSize=(48,48)
    )

    if len(faces) == 0:
        print(f"No faces detected in {img_path}")
    else:
        # c) Annotate each face
        for (x, y, w, h) in faces:
            face_img = gray[y:y+h, x:x+w]
            face_resized = cv2.resize(face_img, (48, 48)).astype('float32') / 255.0
            face_input = np.expand_dims(face_resized, axis=(0, -1))

            # d) Predict emotion
            preds = model.predict(face_input)
            idx = np.argmax(preds[0])
            label = emotions[idx]
            conf = preds[0][idx]

            # e) Print result to console
            print(f"Detected Expression: {label} with confidence {conf:.2f} in {img_path}")

            # f) Draw on image
            cv2.rectangle(frame, (x, y), (x+w, y+h), (0,255,0), 2)
            cv2.putText(
                frame,
                f"{label} ({conf:.2f})",
                (x, y-10),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.8,
                (0,255,0),
                2
            )

    # g) Show and save results
    output = img_path.rsplit('.', 1)
    output_path = output[0] + '_fer.png'
    cv2.imwrite(output_path, frame)
    print(f"Processed {img_path} -> {output_path}")
    cv2.imshow('FER Result', frame)
    key = cv2.waitKey(0)
    if key == ord('q'):
        break

cv2.destroyAllWindows()
```
